```python
import numpy as np
from GMM import EM_GMM, VI_GMM, GS_GMM
import matplotlib.pyplot as plt
%matplotlib inline

X = np.loadtxt('data.txt',delimiter=',')

model = GS_GMM(X,n_iter=500)

model.fit()
```

```python
l = []
for cluster in model.ns_iter:
    observations = np.sort(np.bincount(cluster.astype(int))).tolist()
    l.append(observations[::-1][:6])

array = np.asarray(l)

x=[1,2,3,4,5,6]
for i in range(500):
    y = array[i]
    print y
    plt.scatter(x,y,label=i)
plt.legend()
plt.show()
```

```python
l = []
for cluster in model.clusters_iter:
    l.append(len(np.unique(cluster)))
plt.plot(l)
plt.show()
```
